```python
from pyagadir.models import AGADIR

agadir = AGADIR(pH=7.4, M=0.15, T=298.15)
result = agadir.predict("AAAAAAAAAAAAAAAA", ncap="Z", ccap="B")
print(result.get_percent_helix())
```

```python
def process_data(filepath):
    ph_data = []
    helix_data = []
    with open(filepath, "r") as f:
        for line in f:
            if line.strip() == "":
                continue
            line = line.strip().replace(",", ".")
            ph, helix = line.split("; ")
            ph_data.append(round(float(ph), 2))
            helix_data.append(round(float(helix), 2))
    return ph_data, helix_data

ph_data, helix_data = process_data("4h_predicted.txt")
print(ph_data)
print(helix_data)
```
